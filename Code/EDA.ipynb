# 2. EDA

> 농가수 대비 두수

> 농가수 대비 처리시설 개수

> 읍면동별_농가수 는 EDA용

> 면적대비 파생변수를 만들어 모델링용 변수 생성

```python
plt.figure(figsize=[15, 15])

plt.subplot(211)
sns.barplot(x='시도', y='2019_가축분뇨_농가수_총합', data=df)
plt.xticks(rotation=45)
plt.title('시도별 농가수')
plt.xlabel('')

plt.subplot(212)
sns.barplot(x='시도', y='2019_가축분뇨_두수_총합', data=df)
plt.xticks(rotation=45)
plt.title('시도별 두수')
plt.xlabel('')

plt.show()
```

```python
plt.figure(figsize=[20, 15])

plt.subplot(211)
sns.barplot(x='시도', y='분뇨발생량_총합', data=df)
plt.xticks(rotation=45)
plt.title('시도별 분뇨발생량')
plt.xlabel('')

plt.subplot(223)
sns.barplot(x='시도', y='분뇨처리시설 개수', data=df)
plt.xticks(rotation=45)
plt.title('시도별 분뇨처리시설 개수')
plt.xlabel('')

plt.subplot(224)
sns.barplot(x='시도', y='하수처리시설 개수', data=df)
plt.xticks(rotation=45)
plt.title('시도별 하수처리장 개수')
plt.xlabel('')

plt.show()
```
