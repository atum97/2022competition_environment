```python
import pandas as pd
import numpy as np
import os

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Basic_Dataset_충청경기.xlsx')
df
```

```python
질병 = pd.read_excel('./Data/독립변수_v4/경기도_가축질병_전처리.xlsx')
질병
```

```python
질병1 = 질병[['시도','시군','읍면동','가축전염병명']].groupby(by=['시도','시군','읍면동']).count().reset_index()
질병1
```

```python
질병[['시도','시군','읍면동','감염두수']].groupby(by=['시도','시군','읍면동']).sum().reset_index()['감염두수']
```

```python
질병1['감염두수'] = 질병[['시도','시군','읍면동','감염두수']].groupby(by=['시도','시군','읍면동']).sum().reset_index()['감염두수']
질병1
```

```python
df
```

```python
df = pd.merge(df, 질병1, on=['시도','시군','읍면동'], how='left')
df
```

```python
df.rename(columns={'가축전염병명':'질병감염발생수_count'}, inplace=True)
```
